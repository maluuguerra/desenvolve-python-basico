```python
print("Modulo 3, condicionais")
print("Q1.")

n1 = float(input('Digite o primeiro operando: '))
op = input('Digite o operador: ')
n2 = float(input('Digite o segundo operando: '))
soma = 0
if(op == '+'):
  soma = n1 + n2
elif(op == '-'):
  soma = n1 -  n2
elif(op == '/'):
  soma = n1 / n2
elif(op == '*'):
  soma = n1 * n2
elif(op == '**'):
  soma = n1 ** n2

print("resultado %.1f" %(soma))



```

```python
print("Q2.")
l1 = int(input('Digite valor do lado 1: '))
l2 = int(input('Digite valor do lado 2: '))
l3 = int(input('Digite valor do lado 3: '))

if((l1 == l2 != l3) or (l1 == l3 != l2) or (l2 == l3 != l1) ):
  print("Triângulo: isóceles")
elif (l1 == l2 == l3):
  print("Triângulo Equilátero")
else:
  print("Triângulo: Escaleno")

```

```python
print("Q3.")

pontos = int(input('Informe a pontuação: '))

if(pontos < 70 ):
  print("Insatisfatório")
elif(pontos >= 70 and pontos < 80):
  print("Regular")
elif(pontos >= 80 and pontos < 90):
  print("Bom")
elif(pontos >= 90):
  print("Excelente")

```
