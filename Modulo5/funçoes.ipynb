```python
print("Modulo 5 - Funções ")
print("Q1.")

def fatorial(n):
  fat = 1
  for i in range(1, n+1):
    fat = fat * i
  return fat

x = int(input("Digite n:"))
resultado = fatorial(x)
print(resultado)
```

```python

print("Q2.")

def soma_quadrado(x, y):
  soma = (x**2) + (y**2)
  return soma

a = int(input("Informe o primeiro número:"))
b = int(input("Informe o segundo número:"))

resultado = soma_quadrado(a,b)
print(resultado)

```

```python
print("Q3.")

def soma_digitos(num):
  soma = 0
  while (num !=0):
    novo = num % 10
    num = num // 10
    soma += novo
  return soma

n = int(input("Digite um numero:"))
result = soma_digitos(n)
print(result)
```

```python
print("Q4.")

def inverte_valor(n):
  invertido = 0
  while(n!=0):
    novo = n % 10
    n = n // 10
    invertido = (invertido * 10) + novo
  return invertido

def verifica_invertido (n, inverso):
  if (((n%2)==0) and ((inverso%2)==0)):
    return 1
  elif(((n%2)!=0) and ((inverso%2)!=0)):
    return 2
  else:
    return 0


numero = int(input("Informe um número com pelo menos dois dígitos: "))

inv = inverte_valor(numero)
print(f"O inverso de {numero} é {inv}")

comp = verifica_invertido(numero, inv);

if(comp == 1):
  print("Os números são igualmente pares!")
elif(comp == 2):
  print("Os números são igualmente ímpares!")
else:
  print("Os números são diferentes!")




```

```python
import math
print("Q5.")

def calcula_perimetro_triangulo(a, b, c):
  return (a + b + c)

def calcula_perimetro_circulo(n):
  return(2 * math.pi * n)
def calcula_perimetro_retangulo(l1, l2 = 0) :
  if (l2 == 0):
    return (4 * l1)
  else:
    return ((2 * l2 )+ (2 * l2))


print("1 - Triangulo")
print("2 - Circulo ")
print("3 - Retangulo")
print("4 - sair")

while True:
  perimetro = 0
  op = int(input("Opção: "))
  if(op == 4):
    break

  if(op == 1):
    a = int(input("Informe o primeiro lado:"))
    b = int(input("Informe o segundo lado:"))
    c = int(input("Informe o terceiro lado:"))
    perimetro = calcula_perimetro_triangulo(a, b, c)
    print(f"O perímetro é: {perimetro}")
  elif(op == 2):
    r = int(input("Informe o raio:"))
    perimetro = calcula_perimetro_circulo(r)
    print("O perímetro é: %.2f" %(perimetro))
  else:
    l1 = int(input("Informe o primeiro lado:"))
    l2 = int(input("Informe o segundo lado:"))
    perimetro = calcula_perimetro_retangulo(l1,l2)
    print(f"O perímetro é: {perimetro}")

```
