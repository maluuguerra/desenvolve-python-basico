```python
n = int(input("Digite a quantidade de jogos: "))

vitorias = 0
empates = 0
derrotas = 0

for i in range(n):
    gols_galo = int(input())
    gols_oponente = int(input())


    if gols_galo > gols_oponente:
        vitorias += 1
    elif gols_galo == gols_oponente:
        empates += 1
    else:
        derrotas += 1

pontuacao_total = (vitorias * 3) + (empates * 1)

print(f"Resumo do Campeonato:")
print(f"Vitórias: {vitorias}")
print(f"Empates: {empates}")
print(f"Derrotas: {derrotas}")
print(f"Pontuação Total: {pontuacao_total}")
```

```python
maior = 0
menor = 99999

print("Digite valores inteiros (digite 0 para encerrar):")

while True:
    valor = int(input("Valor: "))

    if valor == 0:
        break

    if (valor > maior):
        maior = valor

    if (valor < menor):
        menor = valor


print(f"maior: {maior}")
print(f"menor: {menor}")
```
